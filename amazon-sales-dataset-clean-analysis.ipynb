# **DATA EXPLORATION**

```python
#Setting Visualization Style

sns.set_style(style='dark')

sns.set_palette(palette="cubehelix")
```

## **Observation 1: Product Category**

```python
amazondf['discount_price'] = pd.to_numeric(amazondf['discount_price'], errors='coerce')
# plot data
fig, ax = plt.subplots(figsize=(12, 8))

# get the median discount price for each category
medians = amazondf.groupby(['product_category'])['discount_price'].median().reset_index().sort_values('discount_price', ascending=False)

# plot data
sns.barplot(ax=ax, data=amazondf, x='product_category', y='discount_price', palette='magma',order=medians['product_category'])

# set y-axis limit to maximum value in data
plt.ylim(0, amazondf['discount_price'].max())

# set plot title and labels
ax.set_title('Prices by Product Category after Discount', fontweight='bold')
ax.set_xlabel('Product Category', fontweight='bold')
ax.set_ylabel('Prices', fontweight='bold')

plt.tight_layout(h_pad=2)

plt.show()
```

```python
category_list = amazondf['product_category'].unique()
```

```python
amazondf['actual_price']=amazondf['actual_price'].str.replace(',', '')
amazondf['actual_price']=amazondf['actual_price'].astype(float)
```

```python
# Group dataframe by product category
grouped = amazondf.groupby('product_category')

# Set up figure and axes
fig, axs = plt.subplots(nrows=len(grouped), figsize=(8, 6 * len(grouped)))

# Loop through each group and create a barplot of top 5 expensive products
for i, (category, group) in enumerate(grouped):
    top_expensive = group.sort_values('actual_price', ascending=False).head(5)
    sns.barplot(ax=axs[i], data=top_expensive, x='actual_price', y='name', palette='magma')
    axs[i].set_title(f'Top 5 Expensive Products in {category}', fontweight='bold')
    axs[i].set_xlabel('Discounted Price', fontweight='bold')
    axs[i].set_ylabel('Product Name', fontweight='bold')
    
plt.tight_layout()
plt.show()
```

```python
#Top 5 Cheapest Products After Discount

cheap_prod = sns.barplot(data=amazondf.sort_values('discount_price').head(5), x='discount_price', y='name')

cheap_prod.set_title('Top 5 Cheapest Products After Discount', fontweight='bold')
cheap_prod.set_xlabel('Discounted Price', fontweight='bold')
cheap_prod.set_ylabel('Product Name', fontweight='bold')

plt.show()
```

## **Observation 2: Correlation between Features**

```python
#Heatmap & Correlation between Actual Price & Discounted Price
figure, ax = plt.subplots(figsize=(10, 8))

# Set the title for the figure
figure.suptitle('Correlation Between Features', fontweight='heavy', size='xx-large')

# Create heatmap in the axis
sns.heatmap(ax=ax, data=amazondf.corr(), annot=True)
ax.set_title('Heatmap', fontweight='bold')

# Adjust the space between the subplots
plt.subplots_adjust(hspace = 0.8)

# Display the plot
plt.show()
```

```python
sns.scatterplot(data=amazondf, x='actual_price', y='discount_price', hue='product_category')
plt.title('Price Comparison')
plt.xlabel('Actual Price')
plt.ylabel('Discounted Price')
plt.show()
```

```python
plt.hist(amazondf['actual_price'], bins=20, color='blue', alpha=0.5)
plt.title('Actual Price Distribution')
plt.xlabel('Actual Price')
plt.ylabel('Count')
plt.show()
```

```python
sns.lineplot(data=amazondf, x='no_of_ratings', y='ratings', hue='product_category')
plt.title('Ratings vs Number of Ratings')
plt.xlabel('Number of Ratings')
plt.ylabel('Ratings')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(12, 8)) # modify figsize as per your preference
sns.boxplot(data=amazondf, x='product_category', y='discount_percentage', ax=ax)
ax.set_title('Discount Percentage Distribution')
ax.set_xlabel('Product Category')
ax.set_ylabel('Discount Percentage')
plt.show()
```

```python
#Discount Percentage Distribution

discounts = sns.histplot(data=amazondf, x='discount_percentage', bins=8, kde=True, color='green')


discounts.set_xlabel('Discount Percentage', fontweight='bold')
discounts.set_ylabel('Count', fontweight='bold')
discounts.set_title('Discount Percentage Distribution', fontweight='heavy', size='x-large')

plt.show()
```

Most products on the dataset have discounts at around **50% - 65%**.
